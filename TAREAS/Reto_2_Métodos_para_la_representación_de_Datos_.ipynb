***Representacion*** ***Estatica***

La estructura tiene un tamaño fijo durante la ejecucion, no cambia dinamicamente.

ESCENARIO: Registro de Estudiantes.

TAREA: Crear un programa donde busque al estudiante por su numero de registro y que lance sus datos (Nombre, carrera en la que estudia y su registro mas).

```python
"""
Registro de Estudiantes(representacion
estatica mejorada).
Permite al usuario agregar
un nuevo estudiante al registro.
"""
#Lista Predefinida de Estudiantes
estudiantes = [  #Usando estudiantes como mi diccionario
    {
        "Nombre" :"Alvaro Azurduy",
        "Carrera" :"Ingenieria De Sistemas",
        "Matricula" :"222007540"
    },
    {
        "Nombre" :"Jose Ayala",
        "Carrera" :"Ingenieria Petrolera",
        "Matricula" :"225008340"
    },
    {
        "Nombre" :"Dennis Melgar",
        "Carrera" :"Medicina",
        "Matricula" :"223014340"
    }
]


def Mostrar_Estudiantes():
  """
  Muestra todos los estudiantes registrados
  """
  print("\n Lista de Estudiantes:\n")
  for estudiante in estudiantes:
    print(f"Nombre : {estudiante['Nombre']}")
    print(f"Carrera : {estudiante['Carrera']}")
    print(f"Matricula : {estudiante['Matricula']}\n")

def Buscar_Estudiante(matricula):
  """
  Busca un estudiante por su matricula.

  Parametros:
  matricula(str): Matricula del Estudiante

  Retorna:
  dict or None: Datos del Estudiante si encuentra.
  Y si no retorna "Que no existe el Estudiante"
  """
  for estudiante in estudiantes:
    if estudiante["Matricula"] == matricula :
      return estudiante
  return None


#------------------------Menu Principal-----------------------------
def menu():
    """
    Menu principal del programa.
    """
    while True:
      print(" \n MENU DEL REGISTRO ")
      print(" 1. Mostrar Estudiantes ")
      print(" 2. Buscar Estudiante ")
      print(" 3. Salir del programa ")

      opcion = input(" Seleccione una Opcion (1-3): ")

      if opcion == "1":
         Mostrar_Estudiantes()
      elif opcion == "2":
        Matricula = input(" Ingrese Matricula a Buscar ")
        resultado = Buscar_Estudiante(Matricula)
        if resultado:
          print("\n Estudiante encontrado ")
          print(f"Nombre: {resultado['Nombre']} ")
          print(f"Carrera: {resultado['Carrera']} ")
          print(f"Matricula: {resultado['Matricula']} ")
        else:
          print("\n Estdiante no encontrado con ese numero de Matricula. ")
      elif opcion == "3":
        print(" Saliendo del programa ")
        break

      else:
        print("Opcion Invalida. Por favor, intente de nuevo")


#Mostrar el Menu
menu()

```

***Representacion*** ***Dinamica***

   La memoria se asigna y libera durante la ejecucion.

   Escenario: Agenda de Contactos.

   Tarea: Realizar un programa donde el usuario realice utilidades sobre su agenda de contactos(agregar contactos, buscar contacto, eliminar, mostrar los contactos).

```python
contactos=[] #Diccionario


def Mostrar_Contactos():
  """
  Muestra todos los contactos registrados.
  """
  if not contactos:
    print("\n No hay contactos en la agenda")
    return
  else:
    print("\n Lista de contactos:")
    for i,c in enumerate(contactos,1):
      print(f"{i}.{c['nombre']-{c['telefono']}}")

def Buscar_Contacto(nombre):
  """
  Busca el contacto por su nombre.

  Parametro:
  nombre(str): Nombre del contacto.

  Retorna:
  dir or None:Nombre y telefono del contacto que se encuentra en la agenda
  y si no, devuelve "Contacto no encontrado"
  """
  encontrado = [c for c in contactos if nombre.lower() in c['nombre'].lower()]
  if encontrado:
    for c in encontrado:
      print(f"-{c['nombre']}:{c['telefono']}")
  else:
    print("No se encontraron contactos con ese nombre!")

def Agregar_Contacto(nombre, telefono):
  """
  Agrega un contacto a la agenda con su nombre y telefono.

  Parametro:
  nombre(str): Nombre del contacto.
  telefono(str): Telefono del contactacto.

  Retorna:
  Contacto registrado a la Agenda.
  """
  contactos.append({"nombre":nombre,"telefono":telefono})

def Eliminar_Contacto(nombre):
  """
  Elimina el contacto de la agenda.

  Parametro:
  nombre(str): Nombre del contacto

  Retorna:
  Contacto eliminado de la agenda.
  """
  if Buscar_Contacto(nombre):
    for c in contactos:
      if c["nombre"].lower() == nombre.lower():
        contactos.remove(c)
        print("Contacto Eliminado con exito")
        return
  else:
    print("El contacto no se encuentra en la agenda.")
    print("No hay nada que eliminar!")

#---------Mostrar el menu---------------
def menu():
  while True:
    print("\n-------AGENDA DE CONTACTOS-------")
    print("1.Agregar Contacto")
    print("2.Mostrar Contactos")
    print("3.Buscar Contacto")
    print("4.Eliminar Contacto")
    print("5.Salir del programa")

    opcion= input("Seleccione una opcion (1-5): ")

    if opcion == "1":
      print("\nAgregando Contacto")
      nombre= input("Nombre:")
      telefono= input("Telefono:")
      Agregar_Contacto(nombre,telefono)
      print("\n Contacto agregado a la agenda.")

    elif opcion == "2":
      Mostrar_Contactos()

    elif opcion == "3":
       nombre= input("\n Nombre del contacto que desea buscar:")
       Buscar_Contacto(nombre)

    elif opcion == "4":
       nombre= input("\n Nombre del contacto que desea eliminar:")
       Eliminar_Contacto(nombre)

    elif opcion == "5":
      print("\n Saliendo de la Agenda")
      break

    else:
      print("\n Opcion invalida, porfavor intentelo de nuevo")

#Mostrar el menu
menu()






```


***RESENTACION*** ***PERSISTENTE***

Los datos se guardan permanentemente, normalmente en archivos o bases de datos.

Escenario: Gestionar una colección de libros favoritos guardados de forma persistente.

Tarea: Permitir agregar, mostrar, buscar y eliminar libros, conservando los datos en un archivo JSON.

```python
# ---------- Gestión de Libros Favoritos ----------

# Cargar libros desde archivo
def cargar_libros(filename="libros.txt"):
    libros = []
    try:
        with open(filename, "r") as file:
            for linea in file:
                titulo, autor = linea.strip().split(" - ")
                libros.append({"titulo": titulo, "autor": autor})
    except FileNotFoundError:
        pass
    return libros

# Guardar libros en archivo
def guardar_libros(libros, filename="libros.txt"):
    with open(filename, "w") as file:
        for libro in libros:
            file.write(f"{libro['titulo']} - {libro['autor']}\n")

# Agregar libro
def agregar_libro(libros, titulo, autor):
    libros.append({"titulo": titulo, "autor": autor})
    print(f"Libro '{titulo}' agregado.")

# Eliminar libro
def eliminar_libro(libros, titulo):
    encontrado = next((l for l in libros if l["titulo"] == titulo), None)
    if encontrado:
        libros.remove(encontrado)
        print(f"Libro '{titulo}' eliminado.")
    else:
        print(f"Libro '{titulo}' no se encuentra en la lista.")

# Mostrar libros
def mostrar_libros(libros):
    if libros:
        print("\nLista de libros:")
        for i, libro in enumerate(libros, 1):
            print(f"{i}. {libro['titulo']} - {libro['autor']}")
    else:
        print("No hay libros en la lista.")

# ---------------- Ejemplo de uso ----------------
libros = cargar_libros()

agregar_libro(libros, "Cien Años de Soledad", "Gabriel García Márquez")
agregar_libro(libros, "1984", "George Orwell")
agregar_libro(libros, "El Principito", "Antoine de Saint-Exupéry")

mostrar_libros(libros)

eliminar_libro(libros, "1984")

mostrar_libros(libros)

guardar_libros(libros)
```

***REPRESENTACION*** ***DE*** ***DATOS*** ***SIMULADOS***

Estructuras de datos que existen solo en memoria y se pierden al cerrar el programa, usadas para pruebas o simulaciones.

Escenario: Simulación del flujo de personas entrando y saliendo de un gimnasio en distintos momentos.

Tarea: Registrar en memoria las personas presentes, permitiendo agregarlas (entrar), eliminarlas (salir) y mostrar la lista actual, sin guardar los datos en archivos.

```python
import random
import time

# ---------- Clases ----------

class Persona:
    def __init__(self, nombre):
        self.nombre = nombre

class Gimnasio:
    def __init__(self):
        self.personas_presentes = []

    def entrar(self, persona):
        self.personas_presentes.append(persona)
        print(f"{persona.nombre} entró al gimnasio.")

    def salir(self, persona):
        if persona in self.personas_presentes:
            self.personas_presentes.remove(persona)
            print(f"{persona.nombre} salió del gimnasio.")

    def mostrar_personas(self):
        if self.personas_presentes:
            print("Personas en el gimnasio:")
            for p in self.personas_presentes:
                print(f"- {p.nombre}")
        else:
            print("No hay personas en el gimnasio.")

# ---------- Simulación ----------

nombres = ["Ana", "Luis", "Carlos", "Marta", "Pedro"]
gim = Gimnasio()
personas = [Persona(nombre) for nombre in nombres]

# Simulación de 5 pasos
for paso in range(5):
    print(f"\n--- Paso {paso+1} ---")
    accion = random.choice(["entrar", "salir"])
    persona = random.choice(personas)

    if accion == "entrar":
        gim.entrar(persona)
    else:
        gim.salir(persona)

    gim.mostrar_personas()
    time.sleep(1)
```
